###Fractionnons les données

```python
#Variables d'entrées
X = data.drop(["ID_Pret","Statut_Pret"], axis = 1) #on retire ici la colonne ID_pret qui
                                                                            #n'est pas utile à notre modèle

#Variable cible
y = data["Statut_Pret"]
```

###Normalisons les varianbvles d'entrées

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
```

###Splittons le jeu de données

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size= 0.8, random_state=42)
```

#Model KNN

###A la recherche du K OPTIMAL

```python
#importation du modele et d'un metric pour mesurer la precision
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
scores = []
for k in range(1,21):
  mon_model_knn = KNeighborsClassifier(n_neighbors=k) #instanciation du modele avec chaque valeur de k
  mon_model_knn.fit(x_train, y_train) #entrainer le modele
  y_pred = mon_model_knn.predict(x_test) #prédiction
  score = accuracy_score(y_test, y_pred) #calcul de la précision du modele
  scores.append(score)

k_optimal = scores.index(max(scores)) + 1 #car le k optimal est celui qui genere le plus grand score de
                                          #precision(max(scores)) et aussi la position de ce max est son index plus 1
k_optimal
```

###Model final

```python
#Réinstanciation du modele
mon_model_knn = KNeighborsClassifier(n_neighbors=k_optimal)
#Réentrainement du modele
mon_model_knn.fit(x_train,y_train)
```
